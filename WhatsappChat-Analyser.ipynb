```python
 def sentimentalAnalysis(data,columnname):
        
        #downloading vader_lexicon for the process
        import matplotlib.pyplot as plt
        import seaborn as sns
        "Importing Necessary Packeage"
        from nltk.sentiment.vader import SentimentIntensityAnalyzer
        sid = SentimentIntensityAnalyzer()
        
        "Deleting null pr empty value"
        data.dropna(inplace=True)
        
        "Checking for a comment"
        #sid.polarity_scores(data[columnname][93])
        
        "Creating respective columns"
        
        data['scores'] = data[columnname].apply(lambda commentText: sid.polarity_scores(commentText))
        data['compound']  = data['scores'].apply(lambda score_dict: score_dict['compound'])
        data['Negtive']  = data['scores'].apply(lambda score_dict: score_dict['neg'])
        data['Postive']  = data['scores'].apply(lambda score_dict: score_dict['pos'])
        data['Neutral']  = data['scores'].apply(lambda score_dict: score_dict['neu'])
        
        "Creating final pos or neg using compound score"
        data['comp_score'] = data['compound'].apply(lambda c: 'pos' if c >=0 else 'neg')
        plt.clf()
        #comp=sns.countplot(x = 'comp_score', hue = 'Name', data = data, palette = 'magma')
        #comp.figure.savefig("date_charts.png")
        "Checking how many pos and neg"
        posneg=pd.DataFrame(data['comp_score'].value_counts())
        return posneg,data
    
```

```python
pos,data_Senti=sentimentalAnalysis(dataset,columnname='Chat')
```

### Topic Modelling - NMF

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf=TfidfVectorizer(max_df=0.95,min_df=2,stop_words='english')
dtm=tfidf.fit_transform(df["Chat"])
```

```python
from sklearn.decomposition import NMF
nmf_model=NMF(n_components=5,random_state=42)
nmf_model.fit(dtm)
```

```python
for index,topic in enumerate(nmf_model.components_):
    results=([tfidf.get_feature_names_out()[i] for i in topic.argsort()[-10:]])
    print(results)
```

```python
topic_results=nmf_model.transform(dtm)
df["Topic"]=topic_results.argmax(axis=1)
```

### Topic Modelling - LDA

```python
from sklearn.decomposition import LatentDirichletAllocation

# Create LDA model
lda_model = LatentDirichletAllocation(
    n_components=5,       # number of topics
    random_state=42,      # for reproducibility
    learning_method='batch'  # or 'online'
)

# Fit LDA model to your document-term matrix
lda_model.fit(dtm)
```

```python
# Get feature names from your vectorizer
feature_names = tfidf.get_feature_names_out()  # or count_vectorizer.get_feature_names_out()

for topic_idx, topic in enumerate(lda_model.components_):
    top_features_idx = topic.argsort()[-10:][::-1]  # Top 10 words for each topic
    top_features = [feature_names[i] for i in top_features_idx]
    print(f"Topic {topic_idx+1}: {', '.join(top_features)}")
```

```python
# Transform the document-term matrix into topic probabilities
doc_topic_dist = lda_model.transform(dtm)  # shape: (n_documents, n_topics)

# Average topic probabilities across all documents
topic_percentages = doc_topic_dist.mean(axis=0) * 100

# Print topic percentages
for i, perc in enumerate(topic_percentages, start=1):
    print(f"Topic {i}: {perc:.2f}%")
```

### Flirt Encounters

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import PassiveAggressiveClassifier,LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report, accuracy_score
```

```python
flirt_df = pd.read_csv("flirt_dataset.csv")
```

```python
flirt_df
```

### Train ML Models

```python
#Train test split
x_train,x_test,y_train,y_test = train_test_split(flirt_df["text"],flirt_df["label"],test_size=0.3,random_state=42)

# TF-IDF vectorization
vectorizer = TfidfVectorizer()
tfidf_train = vectorizer.fit_transform(x_train)
tfidf_test = vectorizer.transform(x_test)

log_reg_model = LogisticRegression()
log_reg_model.fit(tfidf_train,y_train)

y_pred = log_reg_model.predict(tfidf_test)

print("Logistic Regressor Accuracy:", accuracy_score(y_test, y_pred))
print("***************************")
print("\nLogistic Regressor classification report:\n", classification_report(y_test, y_pred))

# Multinomial Naive Bayes
MNB_classifier = MultinomialNB()
MNB_classifier.fit(tfidf_train, y_train)

MNB_pred = MNB_classifier.predict(tfidf_test)

print("MultinomialNB Classifier Accuracy:", accuracy_score(y_test, MNB_pred))
print("**********************************")
print("\nMultipolynomialNB classification report:\n", classification_report(y_test, MNB_pred))

# Passive Aggressive
PAC_classifier = PassiveAggressiveClassifier(max_iter=10, random_state=42)
PAC_classifier.fit(tfidf_train, y_train)

PAC_pred = PAC_classifier.predict(tfidf_test)


print("Passive Aggressive Classifier Accuracy:", accuracy_score(y_test, PAC_pred))
print("**************************************")
print("\nPassive Aggressive Classifier classification report:\n", classification_report(y_test, PAC_pred))
```

```python
flirt_words = ["baby", "darling", "sweetheart","miss you", "cute", "handsome", "pretty", "love you", "dear", "honey"]

def is_flirty(text):
    text = str(text).lower()
    return any(word in text for word in flirt_words)

#  Mark Flirty Messages
df['IsFlirty'] = df['Chat'].apply(is_flirty)

#  Message Count per User
msg_count = df.groupby('Name')['Chat'].count().reset_index(name='MessageCount')

# Most & Least Talkative
most_talkative = msg_count.sort_values(by='MessageCount', ascending=False).iloc[0]
least_talkative = msg_count.sort_values(by='MessageCount', ascending=True).iloc[0]

#  Flirt Percentage per User
flirt_stats = df.groupby('Name')['IsFlirty'].mean().reset_index(name='FlirtPercentage')
flirt_stats['FlirtPercentage'] = flirt_stats['FlirtPercentage'] * 100

final_stats = msg_count.merge(flirt_stats, on='Name')

print("Chat Summary per User")
print(final_stats)
```

```python
print("Flirt Encounter")
print("***************")
print("\n Most Talkative:")
print(f"{most_talkative['Name']} with {most_talkative['MessageCount']} messages")

print("\n Least Talkative:")
print(f"{least_talkative['Name']} with {least_talkative['MessageCount']} messages")

print("\n Flirt Percentage per User:")
for _, row in final_stats.iterrows():
    print(f"{row['Name']}: {row['FlirtPercentage']:.1f}%")
```

### Time Encounters

```python
# Ensure DateTime column is created from Date + Time
df['DateTime'] = pd.to_datetime(df['Date'].astype(str) + ' ' + df['Time'])

# Extract components
df['OnlyDate'] = df['DateTime'].dt.date
df['DayName'] = df['DateTime'].dt.day_name()
df['OnlyTime'] = df['DateTime'].dt.strftime('%I:%M %p')  # 12-hour format

# Most active date
most_active_date = df['OnlyDate'].value_counts().idxmax()

# Most active day of week
most_active_day = df['DayName'].value_counts().idxmax()

# Most active hour (actually specific minute)
most_active_hour = df['OnlyTime'].value_counts().idxmax()

# Average messages per day
avg_msgs_per_day = df.groupby('OnlyDate').size().mean()

# Display
print("Time Encounters:")
print("*****************")
print(f"Most Active Date: {most_active_date.strftime('%d/%m/%y')}")
print(f"Most Active Day: {most_active_day}")
print(f"Most Active Hour: {most_active_hour}")
print(f"Average Number of Message per Day: {avg_msgs_per_day:.1f}")
```
